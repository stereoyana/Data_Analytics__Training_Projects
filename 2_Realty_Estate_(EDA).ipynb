## Предобработка данных

### Обработка стилистичесих ошибок в названии столбцов

```python
# переименование столбцов cityCenters_nearest, studio и open_plan

df = df.rename(columns={'cityCenters_nearest': 'city_centers_nearest', 'studio': 'is_studio', 'open_plan': 'is_open_plan'})
```

```python
# проверка успешности

df.columns
```

### Обработка дубликатов

Столбец `locality_name` имеет строковый тип данных. Посмотрим уникальные значения в нем, чтобы найти возможные дубликаты:

```python
# просмотр всех уникальных значений в столбце

df.locality_name.sort_values().unique()
```

```python
print('Количество уникальных наименований населенных пунктов до обработки:', df.locality_name.nunique())
```

Обнаружены одинаковые по смыслу значения в разном написании. 

Приведем их к одному написанию:

```python
# устранение дублей в названиях с помощью замены

df.locality_name = df.locality_name.str.replace('ё', 'е')
df.locality_name = df.locality_name.str.replace('садоводческое некоммерческое товарищество', 'садовое товарищество')


doubles = 'городской поселок', 'поселок городского типа', 'поселок при железнодорожной станции', 'поселок станции', 'коттеджный поселок'
for row in doubles:
    df.locality_name = df.locality_name.str.replace(row, 'поселок')
```

```python
# проверка успешности

df.locality_name.sort_values().unique()
```

```python
print('Количество уникальных наименований населенных пунктов после обработки:', df.locality_name.nunique())
```

### Обработка пропущенных значений


Определим и изучим пропущенные значения в соотвествии с задачей:

- Для некоторых пропущенных значений можно предположить логичную замену. Например, если человек не указал число балконов — скорее всего, их нет. Такие пропуски правильно заменить на 0.
- Для других типов данных нет подходящего значения на замену. В этом случае правильно оставить эти значения пустыми. Отсутствие значения — тоже важный сигнал, который не нужно прятать. Заполним пропуски, где это уместно.
- Опишем, почему решили заполнить пропуски именно в этих столбцах и как выбрали значения.
- Укажем причины, которые могли привести к пропускам в данных.

```python
# подсчёт пропусков в столбцах

df.isna().sum()
```

В столбце `is_apartment` 20924 пропусков – наибольшая часть базы. 

Остальные заполненные значения True – это апартаменты. Пропуски появились вероятно из-за технологической ошибки (имелся ввиду False, а не NaN) – количество квартир в продаже статистически больше, чем количество апартаментов. 

Поэтому пропущенные значения можно заменить на `False`. 

```python
# замена пропусков методом fillna и проверка успешности

df.is_apartment = df.is_apartment.fillna(False)
print('Пропусков после обработки:', df.is_apartment.isna().sum())
```

В столбцах `parks_around3000` и `ponds_around3000` – по 5518 пропуска. Это картографические данные, заполняемые автоматически. 
Если система не обнаружила ближайших мест, вероятно их количество равно `0`, а пропуски в данных – это технологическая ошибка.  Можно заменить нулями.

В столбце `balcony` отсутствие данных скорее всего подразумевает негативный ответ на вопрос о наличии балконов. Пропуски в данных – человеческий фактор. Их тоже можно заменить на `0`:

```python
# замена пропусков в трёх столбцах на нули

zero_list = ['ponds_around3000', 'parks_around3000', 'balcony']
print('Пропусков до обработки:', df[zero_list].isna().sum())
print()
df[zero_list] = df[zero_list].fillna(0)
print('Пропусков после обработки:', df[zero_list].isna().sum())
```

В столбцах `ceiling_height`, `floors_total`, `living_area`, `kitchen_area` – пропуски вероятно вызваны человеческим фактором (незнание параметров квартиры). 

В столбцах `days_exposition`, `airports_nearest`, `city_centers_nearest`, `parks_nearest`, `ponds_nearest` – картографические данные, ошибки возникли скорее всего по технологическим причинам. 

Многочисленные отсутствующие значения можно попробовать заполнить рассчитанными данными на основе похожих объявлений.

Рассмотрим эти столбцы подробнее:

```python
column_list = ['ceiling_height', 'living_area', 'kitchen_area', 'days_exposition', 'airports_nearest', 'city_centers_nearest', 'parks_nearest', 'ponds_nearest']

df[column_list].describe()
```

Во всех столбцах обнаружены выбросы минимальных и максимальных значений. Расчету медианы такие выбросы не помешают, но в дальнейшем от них лучше избавиться.

1. В столбце `ceiling_height` небольшой разброс значений с медианой 2.65м.

Предположим, что высота потолков зависит от этажности здания – типовые дома часто имеют стандартную высоту потолков. Пропущенные значения заменим на медиану по этой группировке:

```python
print('Пропусков до обработки:', df.ceiling_height.isna().sum())

df.ceiling_height = df.ceiling_height \
    .fillna(df.groupby(['floors_total'])['ceiling_height'] \
            .transform('median'))

print('Пропусков после обработки:', df.ceiling_height.isna().sum())
```

2. В столбце `living_area` разброс значений высокий – от 18 до 42 метров. 

Предположим, что жилая площадь зависит от общей площади квартиры или статуса `is_studio`. Пропущенные значения заменим на медиану по этим группировкам:

```python
print('Пропусков до обработки:', df.living_area.isna().sum())

df.living_area = df.living_area \
    .fillna(df.groupby(['total_area'])['living_area'] \
            .transform('median'))

print('Пропусков после обработки:', df.living_area.isna().sum())
```

Повторим обработку для квартир-студий:

```python
print('Пропусков до обработки:', df.living_area.isna().sum())

df.living_area = df.living_area \
    .fillna(df.groupby(['is_studio'])['living_area'] \
            .transform('median'))

print('Пропусков после обработки:', df.living_area.isna().sum())
```

3. Предположим, что `kitchen_area` зависит от соотношения к общей плошади обычной квартиры или примерно одинаковы для студий. 

Пропущенные значения заменим на медиану по этим группировкам:

```python
print('Пропусков до обработки:', df.kitchen_area.isna().sum())

df.kitchen_area = df.kitchen_area \
    .fillna(df.groupby(['total_area'])['kitchen_area'] \
            .transform('median'))

df.kitchen_area = df.kitchen_area \
    .fillna(df.groupby(['is_studio'])['kitchen_area'] \
            .transform('median'))

print('Пропусков после обработки:', df.kitchen_area.isna().sum())
```

4. Пропущенные значения в столбцах `airports_nearest`, `city_centers_nearest`, `parks_nearest`, `ponds_nearest` заполним медианой по локации.

```python
location_values = ['airports_nearest', 'city_centers_nearest', 'parks_nearest', 'ponds_nearest']
```

```python
# функция принимает название столбца с пропусками и группировку, по которой считаем медиану

def median_filling(column, group):
    for row in df:
        df[column] = df[column].fillna(df.groupby(group)[column].transform('median'))
```

```python
print('Пропусков до обработки:', df[location_values].isna().sum())
```

```python
# медиана близости аэропорта, центра, парков и прудов в зависимости от локации

median_filling(location_values, ['living_area']) 

print('Пропусков после обработки:', df[location_values].isna().sum())
```

5. Рассчитать корректное время `days_exposition` с текущим набором данных довольно сложно – мало параметров, по которым можно построить корреляцию.

### Категория этажа

Посчитаем и добавим в таблицу категорию этажей квартиры (варианты — первый, последний, другой):

```python
# функция для определения категории этажа

def floor_cat(df):
    floors = df['floor']
    floors_total = df['floors_total']
    if floors == 1:
        return 'первый'
    if floors == floors_total:
        return 'последний'
    return 'другой'
```

### Задание: Также выделите факторы, которые влияют на стоимость квартиры (число комнат, этаж, удалённость от центра, дата размещения объявления). 

Сделайте выводы. Отличаются ли они от общих выводов по всему городу?

```python
# функция для сравнения графиков зависимости в центре СПб и в общей выборке

def final_compare(col):
    fig, axes = plt.subplots(1, 2, figsize=(17,5))
    all_df = df_cleaned \
            .pivot_table(index=col, values='square_meter_price', aggfunc=['median']) \
            .plot(ax=axes[0],title='Вся выборка', color='b', legend=False)
    spb_df = spb_center \
            .pivot_table(index=col, values='square_meter_price', aggfunc=['median']) \
            .plot(ax=axes[1],title='Санкт-Петербург', color='g', legend=False)
    return all_df, spb_df
```

```python
# график зависимости количества комнат и цены за кв.м.

final_compare('rooms');
```

9-комнатные квартиры так же резко растут в цене, 11-комнатные чуть меньше. 

По всей выборке 3-комнатные самые дешевые, а в СПб – 6-комнатные.

```python
# график зависимости категории этажа и цены за кв.м.

final_compare('floor_category');
```

```python
# график зависимости удаленности центра и цены за кв.м.

final_compare('city_centers_nearest');
```

Распределение по категории этажа и удаленности от центра похожее.

```python
# график зависимости дня недели размещения объявления

final_compare('weekday');
```

В отличие от всей выборки, в СПб самые дешевые квартиры размещают по вторникам, а самые дорогие – в понедельник и воскресенье.

```python
# график зависимости месяца размещения объявления

final_compare('month');
```

```python
# график зависимости года размещения объявления

final_compare('year');
```

По месяцу и году размещения ситуация похожая.

```python
compare_list = ['rooms', 'floor_category', 'city_centers_nearest', 'weekday', 'month', 'year']

for col in compare_list:
    final_compare(col);
```

- высокие потолки: 3 метра в отличие от стандартных 2.55 метров;

- в отличие от всей выборки, в СПб самые дешевые квартиры размещают по вторникам, а самые дорогие – в понедельник и воскресенье.
